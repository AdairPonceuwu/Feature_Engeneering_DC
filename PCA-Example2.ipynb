```python
from sklearn.datasets import load_digits
import pandas as pd
import numpy as np
```

```python
dataset = load_digits()
dataset.keys()
```

```python
df_nums = pd.DataFrame(dataset.data, columns=dataset.feature_names)
```

```python
df_nums.head()
```

```python
df_nums.describe()
```

```python
X = df_nums
y = dataset.target
```

```python
# for min_max scaling
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
X_scaled
```

```python
from sklearn.model_selection import train_test_split
# Divide data into training and validation subsets
X_train_full, X_valid_full, y_train, y_valid = train_test_split(X, y, train_size=0.7, test_size=0.3,
                                                                random_state=42)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
score = []
model = RandomForestRegressor()
model.fit(X_train_full, y_train)
score.append(model.score(X_valid_full, y_valid))
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA()
```

```python
X_pca = pca.fit_transform(X)
```

```python
X_pca.shape
```

```python
component_names = [f"PC{i+1}" for i in range(X_pca.shape[1])]
X_pca = pd.DataFrame(X_pca, columns=component_names)

X_pca.head()
```

```python
# Divide data into training and validation subsets
X_train_pca, X_valid_pca, y_train_pca, y_valid_pca = train_test_split(X_pca, y, train_size=0.7, test_size=0.3,
                                                                random_state=42)
```

```python
model = RandomForestRegressor()
model.fit(X_train_pca, y_train_pca)
score.append(model.score(X_valid_pca, y_valid_pca))
```

```python
loadings = pd.DataFrame(
    pca.components_.T,  # transpose the matrix of loadings
    columns=component_names,  # so the columns are the principal components
    index=df_nums.columns,  # and the rows are the original features
)
loadings
```

```python
df_nums["new_feature"] = X_pca.PC55 * X_pca.PC61
df_nums["new_feature3"] = X_pca.PC63 * X_pca.PC1
```

```python
X_train, X_valid, y_train_nex, y_valid_nex = train_test_split(df_nums, y, train_size=0.7, test_size=0.3,
                                                                random_state=42)
```

```python
model = RandomForestRegressor()
model.fit(X_train, y_train_nex)
score.append(model.score(X_valid, y_valid_nex))
```
